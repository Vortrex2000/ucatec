```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
import requests
nltk.download('punkt_tab')
```

```python
# 1. Cargar el texto desde un repositorio
url = "https://www.gutenberg.org/files/11/11-0.txt"  # "Alicia en el país de las maravillas"
response = requests.get(url)
text = response.text
```

```python
# Converción a minúsculas y eliminación de caracteres no deseados
text = text.lower()
```

```python
# 2. Tokenizar el texto
tokens = nltk.word_tokenize(text)
print ("Primero 10 tokens:", tokens[: 10])
```

```python
# 3. N-gramas de longitud 2
bigrams = list(ngrams(tokens, 2))
print("Primeros 10 bigramas:")
print(bigrams[:10])
```

```python
# 4. N-gramas de longitud 3
trigrams = list(ngrams(tokens, 3))
print("\nPrimeros 10 trigramas:")
print(trigrams[:10])
```

```python
# 5. Gráfica de frecuencia de bigramas
fdist_bigrams = FreqDist(bigrams)
fdist_bigrams.plot(10, title="Frecuencia de los 10 bigramas más comunes")
```

```python
# 6. Distribución de frecuencias de bigramas
plt.hist(fdist_bigrams.values(), bins=30, color='blue', edgecolor='black')
plt.title("Distribución de frecuencias de bigramas")
plt.xlabel("Frecuencia")
plt.ylabel("Cantidad de bigramas")
plt.show()
```

```python
# 7. Gráfica de los n-gramas más frecuentes
most_common_bigrams = fdist_bigrams.most_common(10)
most_common_bigrams_labels = [' '.join(gram) for gram, freq in most_common_bigrams]
most_common_bigrams_values = [freq for gram, freq in most_common_bigrams]

plt.bar(most_common_bigrams_labels, most_common_bigrams_values, color='green')
plt.title("Top 10 Bigramas Más Frecuentes")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Frecuencia")
plt.show()

```

```python
#N-gramas de longitud 4
quadgrams = list(ngrams(tokens, 4))
print("Primeros 10 cuadrigramas:")
print(quadgrams[:10])
```

```python
# Frecuencia de los cuadrigramas
fdist_quadgrams = FreqDist(quadgrams)
fdist_quadgrams.plot(10, title="Frecuencia de los 10 cuadrigramas más comunes")
```

```python
# Distribución de frecuencias de cuadrigramas
plt.hist(fdist_quadgrams.values(), bins=30, color='purple', edgecolor='black')
plt.title("Distribución de frecuencias de cuadrigramas")
plt.xlabel("Frecuencia")
plt.ylabel("Cantidad de cuadrigramas")
plt.show()
```
