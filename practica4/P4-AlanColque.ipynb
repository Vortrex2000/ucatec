````python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import requests
````

# Cargar el texto del libro

````python
import pandas as pd

# URL del archivo de texto
url = "https://www.gutenberg.org/files/11/11-0.txt"

texto = pd.read_csv(url, header=None, names=["texto"], engine='python', on_bad_lines='skip')
````

# Tokenizar el texto

````python
nltk.download('punkt')
nltk.download('punkt_tab')  # Descargar el paquete de tokenización si no está disponible
tokens = word_tokenize(texto["texto"].iloc[0])
````

# Generar n-gramas

````python
ngramas_2 = list(ngrams(tokens, 2))
ngramas_3 = list(ngrams(tokens, 3))
````

# Mostrar los 10 primeros n-gramas

````python
print(list(ngramas_3)[:10])
````

# Generar n-gramas de longitud 3

````python
ngramas_3 = ngrams(tokens, 3)
````

Tareas

1. Cambia la longitud de los n-gramas a 4 o 5 y observa cómo cambia la salida.

````python
# Generar n-gramas de longitud 4
ngramas_4 = ngrams(tokens, 4)

# Mostrar los 10 primeros n-gramas de longitud 4
print("Primeros 10 n-gramas de longitud 4:")
print(list(ngramas_4)[:10])

# Generar n-gramas de longitud 5
ngramas_5 = ngrams(tokens, 5)

# Mostrar los 10 primeros n-gramas de longitud 5
print("Primeros 10 n-gramas de longitud 5:")
print(list(ngramas_5)[:10])
````

2. Utiliza la función FreqDist de NLTK para calcular la frecuencia de cada n-grama.

````python
# Calcular la frecuencia de los n-gramas de longitud 2
freq_ngramas_2 = nltk.FreqDist(str(ngrama) for ngrama in ngramas_2)

# Mostrar las 10 n-gramas más frecuentes
print("Los 10 n-gramas más frecuentes de longitud 2:")
print(freq_ngramas_2.most_common(10))
````

3. Utiliza la función plot de matplotlib para visualizar la distribución de frecuencias de los n-gramas.

````python
# Gráfica de distribución de frecuencias de n-gramas
plt.figure(figsize=(12, 6))
plt.hist(freq_ngramas_2.values(), bins=10, edgecolor='black')
plt.xlabel("Frecuencia")
plt.ylabel("Cantidad")
plt.title("Distribución de frecuencias de n-gramas de longitud 2")
plt.show()
````
