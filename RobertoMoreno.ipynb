```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
import requests
```

```python
# Descargar los recursos necesarios de NLTK
nltk.download('punkt_tab')   # Descargamos el paquete 'punkt' para la tokenización
```

```python
# Cargando el texto del libro desde un repositorio
url = "https://www.gutenberg.org/files/11/11-0.txt"  # URL de Alicia en el País de las Maravillas
response = requests.get(url)
text = response.text
```

```python
# Tokenización del texto
tokens = word_tokenize(text)
```

```python
# Generando n-gramas de longitud 2
bigrams = list(ngrams(tokens, 2))
print("10 primeros bigramas:")
print(bigrams[:10])
```

```python
# Generar n-gramas de longitud 3
trigrams = list(ngrams(tokens, 3))
print("\n10 primeros trigrams:")
print(trigrams[:10])
```

```python
# Crear gráfica de frecuencia de bigramas
fdist_bigrams = FreqDist(bigrams)
fdist_bigrams.plot(20, title="Frecuencia de los 20 bigramas más comunes")
```

```python
# Crear gráfica de frecuencia de trigrams
fdist_trigrams = FreqDist(trigrams)
fdist_trigrams.plot(20, title="Frecuencia de los 20 trigrams más comunes")
```

```python
# Crear una gráfica de n-gramas más frecuentes
most_common_bigrams = fdist_bigrams.most_common(10)
bigram_labels, bigram_freqs = zip(*most_common_bigrams)
plt.figure(figsize=(10, 6))
plt.barh([' '.join(label) for label in bigram_labels], bigram_freqs, color='skyblue')
plt.title("Bigramas más comunes", fontsize=14)
plt.xlabel("Frecuencia", fontsize=12)
plt.ylabel("Bigramas", fontsize=12)
plt.show()
```

```python
# Cambiar la longitud de los n-gramas a 4 o 5
quadgrams = list(ngrams(tokens, 4))
quintgrams = list(ngrams(tokens, 5))
```

```python
# Frecuencia de n-gramas de longitud 4
fdist_quadgrams = FreqDist(quadgrams)
print("\n10 primeros quadgrams:")
print(quadgrams[:10])
```

```python
# Visualización de la frecuencia de n-gramas de longitud 4
fdist_quadgrams.plot(20, title="Frecuencia de los 20 quadgrams más comunes")
```

```python
# Frecuencia de n-gramas de longitud 5
fdist_quintgrams = FreqDist(quintgrams)
print("\n10 primeros quintgrams:")
print(quintgrams[:10])
```

```python
# Visualización de la frecuencia de n-gramas de longitud 5
fdist_quintgrams.plot(20, title="Frecuencia de los 20 quintgrams más comunes")
```
